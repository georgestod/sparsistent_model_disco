```python
import os
import glob
from tensorflow.python.summary.summary_iterator import summary_iterator

import matplotlib.pylab as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.utils.random import sample_without_replacement
from sklearn.utils import shuffle,resample
from sklearn.linear_model import Ridge


from time import time

%load_ext autoreload
%autoreload 2
```

```python

l_mse  = []
l_reg  = []
l_loss = []
l_IRC = []
l_IRCada = []
l_PoV = []
l_cnum = []
l_cnum_GT = []

for i,value in enumerate(summary_iterator(glob.glob('*.neptune')[0])):
    try:
        if value.summary.value[0].tag=="3_Training/mse":
            l_mse.append([value.step,value.summary.value[0].simple_value])
        if value.summary.value[0].tag=="3_Training/reg":
            l_reg.append([value.step,value.summary.value[0].simple_value])
        if value.summary.value[0].tag=="3_Training/loss":
            l_loss.append([value.step,value.summary.value[0].simple_value])
        
        if value.summary.value[0].tag=="7_IRC/IRC":
            l_IRC.append([value.step,value.summary.value[0].simple_value])
        if value.summary.value[0].tag=="7_IRC/IRC_Adaptive":
            l_IRCada.append([value.step,value.summary.value[0].simple_value])
         
        if value.summary.value[0].tag=="7_IRC/PoV":
            l_PoV.append([value.step,value.summary.value[0].simple_value])
       
            
        if value.summary.value[0].tag=="7_IRC/cond_num_adaptive":
            l_cnum.append([value.step,value.summary.value[0].simple_value])
            

        if value.summary.value[0].tag=="7_IRC/cond_num_adaptive_GT":
            l_cnum_GT.append([value.step,value.summary.value[0].simple_value])

            
        if i<3:
            print(value)
    except:
        print('no')
```

```python
l_mse_test  = []

for i,value in enumerate(summary_iterator(glob.glob('4_Test/mse_test/1/*.neptune')[0])):
    try:
        if value.summary.value[0].tag=='4_Test/mse_test':
            l_mse_test.append([value.step,value.summary.value[0].simple_value])             
    except:
        print('no')
```

```python
plt.plot(np.array(l_loss)[:,0],np.array(l_loss)[:,1])
plt.plot(np.array(l_mse_test)[:,0],np.array(l_mse_test)[:,1])
plt.yscale('log')
```

```python
glob.glob('5_Proba_Selected/probabilities/1'+'/*.neptune')[0]
```

```python
l_probs=[]
for coeff in range(1,37):

    path =glob.glob('5_Proba_Selected/probabilities/'+str(coeff)+'/*.neptune')[0]
    

    #print(path)
    for i,value in enumerate(summary_iterator(path)):
        
        try:
            if value.summary.value[0].tag=="5_Proba_Selected/probabilities":
                l_probs.append([coeff,value.step,value.summary.value[0].simple_value])
                
        except:
            print('no')
m_probs = np.array(l_probs)
m_probs = m_probs.reshape(36,int(m_probs.shape[0]/36),3)
```

```python
subsampling = 1

fig,ax = plt.subplots(3,1,figsize=(10,8),sharex='col')

ax[0].plot(np.array(l_loss)[::subsampling,0],np.array(l_loss)[::subsampling,1],linewidth=3,label='training loss')
ax[0].plot(np.array(l_mse_test)[::subsampling,0],np.array(l_mse_test)[::subsampling,1],linewidth=3,label='testing mse')

#ax[1].plot(np.array(l_IRC)[::subsampling,0],np.array(l_IRC)[::subsampling,1],linewidth=2,label=r'$\Delta(\Theta,T)$')

#ax[1].plot(np.array(l_IRCada)[::subsampling,0],np.array(l_IRCada)[::subsampling,1],linewidth=4,label=r'$\Delta(\tilde{\Theta},T)$')
#ax[1].plot(np.array(l_PoV)[::subsampling,0],np.array(l_PoV)[::subsampling,1],'-k',linewidth=2,label=r'$\Delta(\tilde{\Theta},\hat{T})$')

#ax[1].plot(np.array(l_cnum_GT)[::subsampling,0],np.array(l_cnum_GT)[::subsampling,1],'-r',linewidth=3,label=r'$\tilde{\Theta}_{\hat{T}}$ cond.num.')

for i in range(1,2):
    ax[i].plot([0,m_probs[0,:,1].max()],[1,1],'--k',linewidth=2,alpha=0.8,label='consistency limit')

    
ax[2].plot(m_probs[0,:,1],np.floor(m_probs[:,:,2].sum(axis=0)),linewidth=2,label='# selected terms')

    


#ax[0].set_xlim([2e3,5e4])
ax[0].set_xscale('log')


ax[0].set_yscale('log')
ax[0].set_ylim([1e-4,2e-3])
ax[0].legend(loc='upper right')

ax[1].set_yscale('log')
#ax[1].set_ylim([1e-6,10])
ax[1].legend(loc='upper left',bbox_to_anchor=(1,1))

ax[2].legend(loc='upper right')
ax[2].set_xlabel('epoch')

#ax[3].set_yscale('log')


#plt.savefig("../../figures/KS_modax_metrics.pdf", bbox_inches='tight')
```

```python

import sys
sys.path.append('../../pdeX')
from sparsity_estimators import Adaptive_Lasso_SS,ir_cond,ir_condAL, get_mask_single


theta = np.array(np.load('theta_FIN.npy'),dtype=np.float32).squeeze()
dt    = np.array(np.load('dt_FIN.npy'),dtype=np.float32).squeeze()


theta_ = np.array(theta,dtype=np.float64)
dt_ = np.array(dt,dtype=np.float64)
normed_dt = dt_/np.linalg.norm(dt_,axis=0,keepdims=True)
normed_theta = theta_/np.linalg.norm(theta_,axis=0,keepdims=True)

mask, for_plots= Adaptive_Lasso_SS(normed_theta,normed_dt)
[tau,piT,pSE,alphas, minLambdaSS, active_set,region] = for_plots
```

```python
fig,ax = plt.subplots(1,1,figsize=(4.5,4))

    
piT = 0.9
efp=12

# average of selected variables
q_hat = tau.sum(axis=1)
# verifying on some upper bound on efp
ev_region = (q_hat)**2/((2*piT-1)*theta.shape[1])    
idxS = (ev_region<efp).argmax()   
minLambdaSS = alphas[idxS]
# selecting variables where the efp is respected
active_set = (tau[idxS:,:]>piT).any(axis=0)   

thrs = np.arange(0.51,1,0.001)

E_V_ub = np.stack([(q_hat)**2/((2*piT-1)*theta.shape[1]) for piT in thrs])

c=ax.contour(alphas, thrs, E_V_ub,levels=[0,1,2,3,4,8,12],cmap='copper')
cf=ax.contourf(alphas, thrs, E_V_ub,levels=[0,1,2,3,4,8],cmap='Spectral')
ax.clabel(c, inline=True,fmt='%0.0f', fontsize=15)

ax.plot([alphas.min(),alphas.max()],[0.9,0.9],'-.',color='white',linewidth=3)
ax.text(5e-6,0.93,r'$\pi_{thr}=0.9$',color='white')

d=fig.colorbar(cf, ax=ax)
d.set_label(r'$EV_{max}$')

ax.set_xlabel(r'$\lambda$')#, fontsize=font_size)
ax.set_ylabel(r'selection probability $\Pi_{k}^{\lambda}$')


for term in range(theta.shape[1]): # here 36 is the number of terms in the library
    if active_set[term]:
        al=1
        ls='-'
        lw=3
    else:
        al=0.3
        ls='--'
        lw=2   
    ax.plot(alphas,tau[:,term],linestyle=ls,alpha=al,linewidth=lw)
    ax.set_xscale('log')

ax.text(3e-5,0.8,r'$uu_x$',color='white')
ax.text(1e-4,0.75,r'$u_{xxxx}$',color='white')
ax.text(2e-4,0.8,r'$u_{xx}$',color='white')

plt.savefig("xKS50_SP.pdf", bbox_inches='tight')
```
