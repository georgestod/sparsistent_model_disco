```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
from matplotlib import rc, font_manager

```

```python
sns.set(font_scale=1.5)
sns.set_style("white")

width = 345

tex_fonts = {
    # Use LaTeX to write all text
    "text.usetex": True,
    "font.family": "serif",
    # Use 10pt font in plots, to match 10pt font in document
    "axes.labelsize": 18,
    "font.size": 10,
    # Make the legend/label fonts a little smaller
    "legend.fontsize": 10,
    "xtick.labelsize": 16,
    "ytick.labelsize": 16
}

plt.rcParams.update(tex_fonts)
```

```python
input_data_B = np.load('oldBurgers40/data_input.npy')
recovered_B = np.load('oldBurgers40/recovered.npy')
prop_recovered_B = np.load('xBurgers40/recovered.npy')

input_data_kdv = np.load('oldKdV15/data_input.npy')
recovered_kdv = np.load('oldKdV15/recovered.npy')
prop_recovered_kdv = np.load('xKdV15/recovered.npy')

input_data_NW = np.load('oldNW20/data_input.npy')
recovered_NW = np.load('oldNW20/recovered.npy')
prop_recovered_NW = np.load('xNW20/recovered.npy')

input_data_KS= np.load('oldKS50/data_input.npy')
recovered_KS = np.load('oldKS50/recovered.npy')
prop_recovered_KS = np.load('xKS50/recovered.npy')
mask_KS = np.load('oldKS50/mask.npy')
```

```python
fig,ax = plt.subplots(3,4, figsize=(8,6))



bx=sns.heatmap(input_data_B,ax=ax[0,0],vmin=0,vmax=1,cmap='Spectral',cbar=False)
bx.invert_yaxis()    
bx.set_xticks([])
bx.set_yticks([])
bx.set_ylabel('input data',fontsize=12)
bx.set_title('Burgers',fontsize=12)
bx=sns.heatmap(recovered_B,ax=ax[1,0],vmin=0,vmax=1,cmap='Spectral',cbar=False)
bx.invert_yaxis()    
bx.set_xticks([])
bx.set_yticks([])
bx.set_ylabel('original DeepMod',fontsize=12)
bx=sns.heatmap(prop_recovered_B,ax=ax[2,0],vmin=0,vmax=1,cmap='Spectral',cbar=False)
bx.invert_yaxis()    
bx.set_xticks([])
bx.set_yticks([])
bx.set_ylabel('rAdaLasso+DeepMod',fontsize=12)


bx=sns.heatmap(input_data_kdv,ax=ax[0,1],vmin=0,vmax=1,cmap='Spectral',cbar=False)
bx.invert_yaxis()    
bx.set_xticks([])
bx.set_yticks([])
bx.set_title('KdV',fontsize=12)
bx=sns.heatmap(recovered_kdv,ax=ax[1,1],vmin=0,vmax=1,cmap='Spectral',cbar=False)
bx.invert_yaxis()    
bx.set_xticks([])
bx.set_yticks([])
bx=sns.heatmap(prop_recovered_kdv,ax=ax[2,1],vmin=0,vmax=1,cmap='Spectral',cbar=False)
bx.invert_yaxis()    
bx.set_xticks([])
bx.set_yticks([])



bx=sns.heatmap(input_data_NW,ax=ax[0,2],vmin=0,vmax=1,cmap='Spectral',cbar=False)
bx.invert_yaxis()    
bx.set_xticks([])
bx.set_yticks([])
bx.set_title('NW',fontsize=12)
bx=sns.heatmap(recovered_NW,ax=ax[1,2],vmin=0,vmax=1,cmap='Spectral',cbar=False)
bx.invert_yaxis()    
bx.set_xticks([])
bx.set_yticks([])
bx=sns.heatmap(prop_recovered_NW,ax=ax[2,2],vmin=0,vmax=1,cmap='Spectral',cbar=False)
bx.invert_yaxis()    
bx.set_xticks([])
bx.set_yticks([])


bx=sns.heatmap(input_data_KS,ax=ax[0,3],mask=mask_KS,vmin=-2,vmax=2,cmap=sns.color_palette("rocket_r", as_cmap=True),cbar=False)
bx.invert_yaxis()    
bx.set_xticks([])
bx.set_yticks([])
bx.set_title('KS',fontsize=12)
bx=sns.heatmap(recovered_KS,ax=ax[1,3],vmin=-2,vmax=2,cmap=sns.color_palette("rocket_r", as_cmap=True),cbar=False)
bx.invert_yaxis()    
bx.set_xticks([])
bx.set_yticks([])
bx=sns.heatmap(prop_recovered_KS,ax=ax[2,3],vmin=-2,vmax=2,cmap=sns.color_palette("rocket_r", as_cmap=True),cbar=False)
bx.invert_yaxis()    
bx.set_xticks([])
bx.set_yticks([])

plt.savefig("interpolations.png", bbox_inches='tight')
```
