```python
fig,ax = plt.subplots(1,4,sharex='row',sharey='row',figsize=(12,3))



###########################################################################################
theta = np.load('xBurgers40/theta_FIN.npy')[0]
dt    = np.load('xBurgers40/dt_FIN.npy')[0]
GT = [2,7]
my_title = 'Burgers'
bx=ax[0]



X = theta/np.linalg.norm(theta,axis=0,keepdims=True)
y = dt/np.linalg.norm(dt,axis=0,keepdims=True)

xiR, for_plotsR = Adaptive_Lasso_SS(X,y,n_alphas=20,nI=100,eps=4,randomized = True)
[tauR,piTR,pSER,alphasR, minLambdaSSR, active_setR,regionR] = for_plotsR
not_GT = np.setdiff1d(np.arange(0,theta.shape[1]),np.array(GT)).tolist()

for term in range(theta.shape[1]):
    meanL = tauR
    stdL  = tauR
    
    if term in GT:
        alpha = 1
        ln_style='-'
        colorr = 'red'
        lab = 'relevant'

    else:
        alpha=0.3
        ln_style='--'
        colorr = 'black'
        lab = 'irrelevant'
        
    bx.plot(alphasR,meanL[:,term],linestyle=ln_style,color=colorr,linewidth=2,alpha=alpha,label=lab)        
    bx.set_xscale('log')
bx.set_xlabel(r'$\lambda$')#, fontsize=font_size)
bx.set_ylabel(r'selection prob. $\Pi_{k}^{\lambda}$')

tau=tauR
alphas=alphasR


cx = bx.get_legend_handles_labels()
plt.legend([cx[0][not_GT[0]],cx[0][GT[0]]],[cx[1][not_GT[0]],cx[1][GT[0]]],loc='lower left')

piT = 0.9
efp=3
# average of selected variables
q_hat = tauR.sum(axis=1)
# verifying on some upper bound on efp
ev_region = (q_hat)**2/((2*piT-1)*X.shape[1])    
idxS = (ev_region<efp).argmax()   
minLambdaSS = alphas[idxS]
# selecting variables where the efp is respected
active_set = (tau[idxS:,:]>piT).any(axis=0)   
thrs = np.arange(0.51,1,0.001)
E_V_ub = np.stack([(q_hat)**2/((2*piT-1)*36) for piT in thrs])
c=bx.contour(alphas, thrs, E_V_ub,levels=[0,1,2,3,4,8,12],cmap='copper')
cf=bx.contourf(alphas, thrs, E_V_ub,levels=[0,1,2,3,4,8,12],cmap='Spectral',alpha=0.6)
bx.clabel(c, inline=True,fmt='%0.0f', fontsize=12)
d=fig.colorbar(cf, ax=bx)
#d.set_label(r'$EV_{max}$')
bx.plot([alphas.min(),alphas.max()],[0.9,0.9],'-.',color='blue',linewidth=3,alpha=0.8)
bx.text(0.5*1e-4,0.93,r'$\pi_{thr}=0.9$',color='blue',alpha=0.8)
bx.set_title(my_title, fontsize=12)






###########################################################################################
theta = np.load('xKdV15/theta_FIN.npy')[0]
dt    = np.load('xKdV15/dt_FIN.npy')[0]
GT = [3,7]
my_title = 'KdV'
bx=ax[1]


#############
X = theta/np.linalg.norm(theta,axis=0,keepdims=True)
y = dt/np.linalg.norm(dt,axis=0,keepdims=True)

xiR, for_plotsR = Adaptive_Lasso_SS(X,y,n_alphas=20,nI=100,eps=4,randomized = True)
[tauR,piTR,pSER,alphasR, minLambdaSSR, active_setR,regionR] = for_plotsR
not_GT = np.setdiff1d(np.arange(0,theta.shape[1]),np.array(GT)).tolist()

for term in range(theta.shape[1]):
    meanL = tauR
    stdL  = tauR
    
    if term in GT:
        alpha = 1
        ln_style='-'
        colorr = 'red'
        lab = 'relevant'

    else:
        alpha=0.3
        ln_style='--'
        colorr = 'black'
        lab = 'irrelevant'
        
    bx.plot(alphasR,meanL[:,term],linestyle=ln_style,color=colorr,linewidth=2,alpha=alpha,label=lab)        
    bx.set_xscale('log')
bx.set_xlabel(r'$\lambda$')#, fontsize=font_size)
#bx.set_ylabel(r'selection prob. $\Pi_{k}^{\lambda}$')

tau=tauR
alphas=alphasR


cx = bx.get_legend_handles_labels()
#plt.legend([cx[0][not_GT[0]],cx[0][GT[0]]],[cx[1][not_GT[0]],cx[1][GT[0]]],loc='lower left')

piT = 0.9
efp=3
# average of selected variables
q_hat = tauR.sum(axis=1)
# verifying on some upper bound on efp
ev_region = (q_hat)**2/((2*piT-1)*X.shape[1])    
idxS = (ev_region<efp).argmax()   
minLambdaSS = alphas[idxS]
# selecting variables where the efp is respected
active_set = (tau[idxS:,:]>piT).any(axis=0)   
thrs = np.arange(0.51,1,0.001)
E_V_ub = np.stack([(q_hat)**2/((2*piT-1)*36) for piT in thrs])
c=bx.contour(alphas, thrs, E_V_ub,levels=[0,1,2,3,4,8,12],cmap='copper')
cf=bx.contourf(alphas, thrs, E_V_ub,levels=[0,1,2,3,4,8,12],cmap='Spectral',alpha=0.6)
bx.clabel(c, inline=True,fmt='%0.0f', fontsize=12)
d=fig.colorbar(cf, ax=bx)
#d.set_label(r'$EV_{max}$')
bx.plot([alphas.min(),alphas.max()],[0.9,0.9],'-.',color='blue',linewidth=3,alpha=0.8)
bx.text(0.5*1e-4,0.93,r'$\pi_{thr}=0.9$',color='blue',alpha=0.8)
bx.set_title(my_title, fontsize=12)




###########################################################################################
theta = np.load('xNW20/theta_FIN.npy')[0]
dt    = np.load('xNW20/dt_FIN.npy')[0]
GT = [0,2,6,18]
my_title = 'NW'
bx=ax[2]


#############
X = theta/np.linalg.norm(theta,axis=0,keepdims=True)
y = dt/np.linalg.norm(dt,axis=0,keepdims=True)

xiR, for_plotsR = Adaptive_Lasso_SS(X,y,n_alphas=20,nI=100,eps=4,randomized = True)
[tauR,piTR,pSER,alphasR, minLambdaSSR, active_setR,regionR] = for_plotsR
not_GT = np.setdiff1d(np.arange(0,theta.shape[1]),np.array(GT)).tolist()

for term in range(theta.shape[1]):
    meanL = tauR
    stdL  = tauR
    
    if term in GT:
        alpha = 1
        ln_style='-'
        colorr = 'red'
        lab = 'relevant'

    else:
        alpha=0.3
        ln_style='--'
        colorr = 'black'
        lab = 'irrelevant'
        
    bx.plot(alphasR,meanL[:,term],linestyle=ln_style,color=colorr,linewidth=2,alpha=alpha,label=lab)        
    bx.set_xscale('log')
bx.set_xlabel(r'$\lambda$')#, fontsize=font_size)
#bx.set_ylabel(r'selection prob. $\Pi_{k}^{\lambda}$')

tau=tauR
alphas=alphasR


cx = bx.get_legend_handles_labels()
#plt.legend([cx[0][not_GT[0]],cx[0][GT[0]]],[cx[1][not_GT[0]],cx[1][GT[0]]],loc='lower left')

piT = 0.9
efp=3
# average of selected variables
q_hat = tauR.sum(axis=1)
# verifying on some upper bound on efp
ev_region = (q_hat)**2/((2*piT-1)*X.shape[1])    
idxS = (ev_region<efp).argmax()   
minLambdaSS = alphas[idxS]
# selecting variables where the efp is respected
active_set = (tau[idxS:,:]>piT).any(axis=0)   
thrs = np.arange(0.51,1,0.001)
E_V_ub = np.stack([(q_hat)**2/((2*piT-1)*36) for piT in thrs])
c=bx.contour(alphas, thrs, E_V_ub,levels=[0,1,2,3,4,8,12],cmap='copper')
cf=bx.contourf(alphas, thrs, E_V_ub,levels=[0,1,2,3,4,8,12],cmap='Spectral',alpha=0.6)
bx.clabel(c, inline=True,fmt='%0.0f', fontsize=12)
d=fig.colorbar(cf, ax=bx)
#d.set_label(r'$EV_{max}$')
bx.plot([alphas.min(),alphas.max()],[0.9,0.9],'-.',color='blue',linewidth=3,alpha=0.8)
bx.text(0.5*1e-4,0.93,r'$\pi_{thr}=0.9$',color='blue',alpha=0.8)
bx.set_title(my_title, fontsize=12)



###########################################################################################
theta = np.load('xKS50/theta_FIN.npy')[0]
dt    = np.load('xKS50/dt_FIN.npy')[0]
GT = [2,4,7]
my_title = 'KS'
bx=ax[3]


#############
X = theta/np.linalg.norm(theta,axis=0,keepdims=True)
y = dt/np.linalg.norm(dt,axis=0,keepdims=True)

xiR, for_plotsR = Adaptive_Lasso_SS(X,y,n_alphas=20,nI=100,eps=4,randomized = True)
[tauR,piTR,pSER,alphasR, minLambdaSSR, active_setR,regionR] = for_plotsR
not_GT = np.setdiff1d(np.arange(0,theta.shape[1]),np.array(GT)).tolist()

for term in range(theta.shape[1]):
    meanL = tauR
    stdL  = tauR
    
    if term in GT:
        alpha = 1
        ln_style='-'
        colorr = 'red'
        lab = 'relevant'

    else:
        alpha=0.3
        ln_style='--'
        colorr = 'black'
        lab = 'irrelevant'
        
    bx.plot(alphasR,meanL[:,term],linestyle=ln_style,color=colorr,linewidth=2,alpha=alpha,label=lab)        
    bx.set_xscale('log')
bx.set_xlabel(r'$\lambda$')#, fontsize=font_size)
#bx.set_ylabel(r'selection prob. $\Pi_{k}^{\lambda}$')

tau=tauR
alphas=alphasR


cx = bx.get_legend_handles_labels()
#plt.legend([cx[0][not_GT[0]],cx[0][GT[0]]],[cx[1][not_GT[0]],cx[1][GT[0]]],loc='lower left')

piT = 0.9
efp=3
# average of selected variables
q_hat = tauR.sum(axis=1)
# verifying on some upper bound on efp
ev_region = (q_hat)**2/((2*piT-1)*X.shape[1])    
idxS = (ev_region<efp).argmax()   
minLambdaSS = alphas[idxS]
# selecting variables where the efp is respected
active_set = (tau[idxS:,:]>piT).any(axis=0)   
thrs = np.arange(0.51,1,0.001)
E_V_ub = np.stack([(q_hat)**2/((2*piT-1)*36) for piT in thrs])
c=bx.contour(alphas, thrs, E_V_ub,levels=[0,1,2,3,4,8,12],cmap='copper')
cf=bx.contourf(alphas, thrs, E_V_ub,levels=[0,1,2,3,4,8,12],cmap='Spectral',alpha=0.6)
bx.clabel(c, inline=True,fmt='%0.0f', fontsize=12)
d=fig.colorbar(cf, ax=bx)
d.set_label(r'$EV_{max}$')
bx.plot([alphas.min(),alphas.max()],[0.9,0.9],'-.',color='blue',linewidth=3,alpha=0.8)
bx.text(0.5*1e-4,0.93,r'$\pi_{thr}=0.9$',color='blue',alpha=0.8)
bx.set_title(my_title, fontsize=12)

plt.savefig("stab_plots_final.pdf", bbox_inches='tight')
```
